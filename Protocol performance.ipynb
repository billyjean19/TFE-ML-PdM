```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
batch_size = [1,10,20,50,100,250]
http_v1 = [25,475,928,1526,3064,7230]
http_v2 = [28,414,688,1881,3490,8684]
mqtt_v1 = [0.7,8.71,16.86,48,120.29,220]
mqtt_v2 = [2,18,39,93,245,528.8]
udp_v1 = [0.93,9.08,19.6,50.9,106.5,256.8]
udp_v2 = [1.28,9.89,19.5,46.8,138.5,299.5]
```

```python
# Plot time series data for Single core
plt.figure(figsize=(12, 6))
plt.plot(batch_size, http_v1, label='HTTP Single-core')
plt.plot(batch_size, mqtt_v1, label='MQTT Single-core')
plt.plot(batch_size, udp_v1, label='UDP Single-core')
plt.xlabel('Batch size')
plt.ylabel('Time to send data [ms]')
plt.legend()
plt.grid(True)
plt.title('Transmission performances')
plt.savefig('single_core.png', format='png')
plt.show()
```

```python
# Plot time series data for Single core (log) 
plt.figure(figsize=(12, 6))
plt.semilogy(batch_size, http_v1, label='HTTP Single-core')
plt.semilogy(batch_size, mqtt_v1, label='MQTT Single-core')
plt.semilogy(batch_size, udp_v1, label='UDP Single-core')
plt.xlabel('Batch size')
plt.ylabel('Time to send data [log ms]')
plt.legend()
plt.grid(True)
plt.title('Transmission performances')
plt.savefig('single_core_log.png', format='png')
plt.show()
```

```python
# Plot time series data for Dual core
plt.figure(figsize=(12, 6))
plt.plot(batch_size, http_v2, label='HTTP Dual-core')
plt.plot(batch_size, mqtt_v2, label='MQTT Dual-core')
plt.plot(batch_size, udp_v2, label='UDP Dual-core')
plt.xlabel('Batch size')
plt.ylabel('Time to send data [ms]')
plt.legend()
plt.grid(True)
plt.title('Transmission performances')
plt.savefig('dual_core.png', format='png')
plt.show()
```

```python
# Plot time series data for Dual core (log)
plt.figure(figsize=(12, 6))
plt.semilogy(batch_size, http_v2, label='HTTP Dual-core')
plt.semilogy(batch_size, mqtt_v2, label='MQTT Dual-core')
plt.semilogy(batch_size, udp_v2, label='UDP Dual-core')
plt.xlabel('Batch size')
plt.ylabel('Time to send data [log ms]')
plt.legend()
plt.grid(True)
plt.title('Transmission performances')
plt.savefig('dual_core_log.png', format='png')
plt.show()
```
